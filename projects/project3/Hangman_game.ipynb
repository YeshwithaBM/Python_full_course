```python
import random
import hangman_stages
import words

print("""Let's Play Hangman!!!!
You have only 6 lives, so try to guess the word within 6 attempts!
Good Luck!
""")

word = random.choice(words.word_list)

hidden_word = []

for _ in range(len(word)):
    hidden_word.append("_")

life = 6

while life > 0:

    print("\nWord:", " ".join(hidden_word))

    guess = input("Guess a letter: ").lower()

    if guess in hidden_word:
        print(f"You've already guessed '{guess}'!")
        continue

    if guess in word:

        for i in range(len(word)):
            if word[i] == guess:
                hidden_word[i] = guess

        if "_" not in hidden_word:
            print("\nWord:", " ".join(hidden_word))
            print("🎉 You Win!")
            break

    else:

        life -= 1

        print("Wrong Guess!")
        print(f"Lives Left: {life}")

        print(hangman_stages.hangman_stages[6 - life])

        if life == 0:
            print("\n💀 You Lose!")
            print("The word was:", word)
```
