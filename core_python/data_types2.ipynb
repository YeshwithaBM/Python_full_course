#### String Slicing

- **Syntax:**

  ```python
  string[start : stop : step]
  ```

- **start** → Index from where slicing begins (**inclusive**).
- **stop** → Index where slicing ends (**exclusive**).
- **step** → Number of positions to move after each character (**optional**).

#### Rules

- Indexing starts from **0**.
- Negative indexing starts from **-1** (last character).
- If `start` is omitted, slicing starts from the beginning of the string.
- If `stop` is omitted, slicing continues until the end of the string.
- The default value of `step` is **1**.
- A negative `step` slices the string in reverse.

````python
text = "Python Programming"

print("Original String:", text)

# Basic Slicing
print("\n--- Basic Slicing ---")
print(text[0:6])      # Python
print(text[7:18])     # Programming
print(text[:6])       # From beginning
print(text[7:])       # Till end
print(text[:])        # Entire string

# Using Step
print("\n--- Using Step ---")
print(text[::2])      # Every 2nd character
print(text[::3])      # Every 3rd character
print(text[1::2])     # Every 2nd character starting from index 1

# Negative Indexing
print("\n--- Negative Indexing ---")
print(text[-11:])     # Programming
print(text[:-12])     # Python
print(text[-6:-1])    # ammin

# Reverse String
print("\n--- Reverse String ---")
print(text[::-1])     # Reverse entire string

# Reverse with Range
print("\n--- Reverse with Range ---")
print(text[5:1:-1])   # noht
print(text[-1:-12:-1])# gnimmargorP

# Copy String
print("\n--- Copy String ---")
copy_text = text[:]
print(copy_text)

# First and Last Characters
print("\n--- First and Last Characters ---")
print(text[:1])       # First character
print(text[-1:])      # Last character

# Last N Characters
print("\n--- Last N Characters ---")
print(text[-5:])      # Last 5 characters

# Excluding Characters
print("\n--- Excluding Characters ---")
print(text[:-1])      # Exclude last character
print(text[1:])       # Exclude first character
print(text[1:-1])     # Exclude first and last characters
````

## String Methods

- String methods are **built-in functions** that perform operations on strings.
- Since strings are **immutable**, string methods do **not modify** the original string.
- Most string methods return a **new string**.
- String methods are called using the **dot (`.`) operator**.

**Syntax:**

```python
string.method_name(arguments)
```

### Common String Methods

- `upper()` → Converts all characters to uppercase.
- `lower()` → Converts all characters to lowercase.
- `title()` → Converts the first letter of each word to uppercase.
- `capitalize()` → Converts the first character of the string to uppercase.
- `swapcase()` → Swaps uppercase letters to lowercase and vice versa.
- `strip()` → Removes leading and trailing whitespace.
- `lstrip()` → Removes leading whitespace.
- `rstrip()` → Removes trailing whitespace.
- `replace(old, new)` → Replaces occurrences of a substring.
- `find(substring)` → Returns the index of the first occurrence or `-1` if not found.
- `index(substring)` → Returns the index of the first occurrence or raises an error if not found.
- `count(substring)` → Counts the occurrences of a substring.
- `startswith(prefix)` → Returns `True` if the string starts with the given prefix.
- `endswith(suffix)` → Returns `True` if the string ends with the given suffix.
- `split(separator)` → Splits the string into a list.
- `join(iterable)` → Joins elements of an iterable into a string.
- `isalpha()` → Returns `True` if all characters are alphabets.
- `isdigit()` → Returns `True` if all characters are digits.
- `isalnum()` → Returns `True` if all characters are alphanumeric.
- `isspace()` → Returns `True` if all characters are whitespace.
- `islower()` → Returns `True` if all characters are lowercase.
- `isupper()` → Returns `True` if all characters are uppercase.

````python
text = "  hello Python Programming  "

print("Original String:", text)

# Case Conversion
print("\n--- Case Conversion ---")
print(text.upper())          # HELLO PYTHON PROGRAMMING
print(text.lower())          # hello python programming
print(text.title())          # Hello Python Programming
print(text.capitalize())     # Hello python programming
print(text.swapcase())       # HELLO pYTHON pROGRAMMING

# Remove Whitespaces
print("\n--- Strip Methods ---")
print(text.strip())          # Removes leading and trailing spaces
print(text.lstrip())         # Removes leading spaces
print(text.rstrip())         # Removes trailing spaces

# Replace
print("\n--- Replace ---")
print(text.replace("Python", "Java"))

# Find, Index, Count
print("\n--- Find, Index & Count ---")
print(text.find("Python"))   # Returns index
print(text.index("Python"))  # Returns index
print(text.count("o"))        # Counts occurrences

# Startswith & Endswith
print("\n--- Startswith & Endswith ---")
clean_text = text.strip()
print(clean_text.startswith("hello"))
print(clean_text.endswith("Programming"))

# Split
print("\n--- Split ---")
words = clean_text.split()
print(words)

# Join
print("\n--- Join ---")
print("-".join(words))

# Checking Methods
print("\n--- Checking Methods ---")
print("Python".isalpha())      # True
print("12345".isdigit())       # True
print("Python123".isalnum())   # True
print("   ".isspace())         # True
print("python".islower())      # True
print("PYTHON".isupper())      # True

# Immutability
print("\n--- Original String ---")
print(text)
````

## String Concatenation

- **String concatenation** is the process of joining two or more strings into a single string.
- The **`+` operator** is used to concatenate strings.
- Only strings can be concatenated with other strings. Concatenating a string with a different data type (such as `int` or `float`) results in a `TypeError`.
- The **`+=` operator** can be used to append one string to another.

**Syntax:**

```python
new_string = string1 + string2
```




````python

first_name = "John"
last_name = "Doe"

full_name = first_name + " " + last_name
print(full_name)
````

#### String Formatting

- **String formatting** is used to insert variables or values into a string.
- It helps create dynamic and readable output.
- The most commonly used method is **f-string**.

**Syntax:**

```python

```


````python

name = "Alice"
age = 22

print(f"My name is {name} and I am {age} years old.")
````

- Python provides **built-in documentation** to help understand modules, functions, classes, and methods.
- The **`help()`** function is used to display the documentation of an object.
- The **`__doc__`** attribute returns the documentation string (docstring) of an object.
