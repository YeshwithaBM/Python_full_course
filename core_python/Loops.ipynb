#### for loop
- A for loop is used to iterate (repeat or traverse) over an iterable., so iterable means traversing a sequence one by one . here sequence collection of items or sequence data types such as list ,tuples,sets,string,dictionary.
- its more like an iterator.
- syntax is : **for var_name in sequence: \n statement(s)**

### For-else 
- else is used only if for loop executed completely successfully. without any break or error .
- The `else` block executes only if the `for` loop finishes without a `break`.
- If `break` is executed, the `else` block is skipped.
- `continue` does **not** skip the `else` block.
- Commonly used in searching and validation problems.

```text
Start
  │
  ▼
Run the for loop
  │
  ▼
Was 'break' executed?
  │
┌─┴─────────┐
│ Yes       │ No
▼           ▼
Skip else   Execute else
  │           │
  └──────► End
```

````python
numbers=[1,2,2,4,32,3]
for i in numbers:
    print(i)
else:
    print("successfully completed")
````

````python
numbers=[1,2,2,4,32,3]
for i in numbers:
    print(i)
    if i==0:
        break
else:
    print("successfully completed")
````

````python
numbers=[1,2,0,4,32,3]
for i in numbers:
    print(i)
    if i==0:
        break
else:
    print("successfully completed") # else is not executed
````

````python
tuple1=(2,56,4,3,5,-1)
for i in tuple1:
    print(i)
    if i==5:
        break
else:
    print("loop success")
print('out of for/else')
````

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
else:
    print("Loop finished")
````

````python
nums = [5, 10, 15, 20]
target = 12

for num in nums:
    if num == target:
        print("Element found")
        break
else:
    print("Element not found")
````

#### `range()` Function

The `range()` function generates a sequence of numbers. It is commonly used with `for` loops to execute a block of code a specific number of times.

- Generates numbers on demand (does not create a list).
- The **start** value is included.
- The **stop** value is excluded.
- The default step size is `1`.

**Syntax**
```python
range(stop) #mandatory to specify
range(start, stop) #optional by default start is 0
range(start, stop, step) #optional, by default step is 1
```

````python
a=range(5)
print(a[0])
print(a)
````
