### For-else 
- else is used only if for loop executed completely successfully. without any break or error .
- The `else` block executes only if the `for` loop finishes without a `break`.
- If `break` is executed, the `else` block is skipped.
- `continue` does **not** skip the `else` block.
- Commonly used in searching and validation problems.

```text
Start
  │
  ▼
Run the for loop
  │
  ▼
Was 'break' executed?
  │
┌─┴─────────┐
│ Yes       │ No
▼           ▼
Skip else   Execute else
  │           │
  └──────► End
```

````python
numbers=[1,2,2,4,32,3]
for i in numbers:
    print(i)
else:
    print("successfully completed")
````

````python
numbers=[1,2,2,4,32,3]
for i in numbers:
    print(i)
    if i==0:
        break
else:
    print("successfully completed")
````

````python
tuple1=(2,56,4,3,5,-1)
for i in tuple1:
    print(i)
    if i==5:
        break
else:
    print("loop success")
print('out of for/else')
````

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
else:
    print("Loop finished")
````

#### `range()` Function

The `range()` function generates a sequence of numbers. It is commonly used with `for` loops to execute a block of code a specific number of times.

- Generates numbers on demand (does not create a list).
- The **start** value is included.
- The **stop** value is excluded.
- The default step size is `1`.

**Syntax**
```python
range(stop) #mandatory to specify
range(start, stop) #optional by default start is 0
range(start, stop, step) #optional, by default step is 1
```

### While loop
- while loop is used to repeat a block of code until given condition is true 
- A `while` loop repeatedly executes a block of code **as long as a given condition is `True`**.

- Used when the number of iterations is **unknown**.
- The condition is checked **before** each iteration.
- The loop stops when the condition becomes `False`.

**Syntax**

```python
while condition:
    # code block
```
```text
Start
  │
  ▼
Check Condition
  │
┌─┴─────────┐
│ True      │ False
▼           ▼
Execute     End
Code
│
└──────────► Check Condition
```

````python
count=1
while count<=5:
    print(count)
    count+=1
````

### Difference Between `for` and `while`

- Use a **`for` loop** when the number of iterations is known or when iterating over a sequence (list, string, tuple, dictionary, range, etc.).
- Use a **`while` loop** when the number of iterations is unknown and execution depends on a condition becoming `False`.

**Examples:**
- `for`: Processing all files in a folder, sending emails to all users, iterating through a list.
- `while`: Waiting for user login, checking server availability, reading input until the user exits.
----

#### Example for **while** Loop 
##### Password Login
- Imagine you're building the login page for an application.
- The user should keep trying until they enter the correct password.

````python
correct_password = "python123"

password = input("Enter Password: ")

while password != correct_password:
    print("Wrong Password! Try Again.")
    password = input("Enter Password: ")

print("Login Successful!")
#The loop continues until the condition becomes false.
````

#### Example for **FOR** loop
##### Sending Emails (for)
- Suppose your company has 5 users.
- You need to send an email to every user exactly once.


````python
users = [
    "Alice",
    "Bob",
    "Charlie",
    "David",
    "Emma"
]

for user in users:
    print(f"Sending email to {user}...")

print("All emails sent successfully!")
#We already have the list.
#We know exactly how many users there are.
#Each user should be processed once.
````
