### Hangman Game Project-3

```python
import random
Start="""Lets Play Hangman!!!!
you have only 6 lives so try to guess the word within 6 attempts! Good luck!"""
print(Start)
word_list = ["tiger", "elephant", 
"rabbit", "monkey", "giraffe", "zebra", "penguin", 
"dolphin", "apple", "banana", "orange", "mango", "grapes", 
"papaya", "cherry", "peach", "python", "laptop", "keyboard",
"monitor", "internet", "database", "software", "computer",
"mountain", "river", "forest", "desert", "ocean", "flower",
"sunshine", "rainbow", "bottle", "window", "pencil",
"backpack", "camera", "wallet", "umbrella", "telephone"]
hangman_stages = [
"""
  +---+
  |   |
  O   |
      |
      |
      |
=========
""",
"""
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
""",
"""
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========
""",
"""
  +---+
  |   |
  O   |
 /|\\ |
      |
      |
=========
""",
"""
  +---+
  |   |
  O   |
 /|\\  |
 /    |
      |
=========
""",
"""
  +---+
  |   |
  O   |
 /|\\  |
 / \\  |
      |
=========
"""
]
word=random.choice(word_list)
hided_word=[]
for i in word:
    hided_word.append("_")
print(hided_word)
guess=input("guess a letter: ")
life=6
while life<=0:
    if "_" not in hided_word:
         print("you win")
         break
    elif guess in word:
      for i in range(len(word)):
        if word[i] == guess:
          hided_word[i] = guess
      print(hided_word)
      guess=input("guess a letter: ")
    else:
       fig=len(hangman_stages)-life   
       print(hangman_stages[fig]) 
       life-=1
       if life==0:
         print("you lose")
       guess=input("guess a letter: ")
```
