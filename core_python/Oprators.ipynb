#### REPL

- REPL(read evaluate print loop) REPL is a python interactive shell.

####  operators 
- Python divides the operators in the following groups:

- Arithmetic operators 
- Assignment operators
- Comparison operators
- Logical operators
- Identity operators
- Membership operators
- Bitwise operators


```python
# arthimetic operators 
a = 10
b = 3

print("Addition:", a + b)          # 13
print("Subtraction:", a - b)       # 7
print("Multiplication:", a * b)    # 30
print("Division:", a / b)          # 3.3333333333333335
print("Floor Division:", a // b)   # 3
print("Modulus:", a % b)           # 1
print("Exponent:", a ** b)         # 1000
```

```python
#assignment operator 
x = 10

print("Initial Value:", x)

x += 5
print("After += :", x)

x -= 3
print("After -= :", x)

x *= 2
print("After *= :", x)

x /= 4
print("After /= :", x)

x //= 2
print("After //= :", x)

x %= 3
print("After %= :", x)

x **= 2
print("After **= :", x)
```

```python
#comparision operators 
a = 10
b = 5

print(a == b)   # False
print(a != b)   # True
print(a > b)    # True
print(a < b)    # False
print(a >= b)   # True
print(a <= b)   # False
```

```python
#logical operators 

a = 10
b = 5

print(a > b and a < 20)   # True
print(a < b and a < 20)   # False

print(a > b or a < 5)     # True
print(a < b or a < 5)     # False

print(not(a > b))         # False
print(not(a < b))         # True
```
