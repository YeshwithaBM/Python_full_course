####  operators 
- Python divides the operators in the following groups:

#### arthimetic operators 

```python

a = 10
b = 3

print("Addition:", a + b)          # 13
print("Subtraction:", a - b)       # 7
print("Multiplication:", a * b)    # 30
print("Division:", a / b)          # 3.3333333333333335
print("Floor Division:", a // b)   # 3
print("Modulus:", a % b)           # 1
print("Exponent:", a ** b)         # 1000
```

#### Assignment Operator

```python
#assignment operator 
x = 10

print("Initial Value:", x)

x += 5
print("After += :", x)

x -= 3
print("After -= :", x)

x *= 2
print("After *= :", x)

x /= 4
print("After /= :", x)

x //= 2
print("After //= :", x)

x %= 3
print("After %= :", x)

x **= 2
print("After **= :", x)
```

#### Comparision Operator

```python
#comparision operators 
a = 10
b = 5

print(a == b)   # False
print(a != b)   # True
print(a > b)    # True
print(a < b)    # False
print(a >= b)   # True
print(a <= b)   # False
```

#### Logical operator

```python
#logical operators 

a = 10
b = 5

print(a > b and a < 20)   # True
print(a < b and a < 20)   # False

print(a > b or a < 5)     # True
print(a < b or a < 5)     # False

print(not(a > b))         # False
print(not(a < b))         # True
```

#### Identity operator

```python
# Memory manager in python reuses the object insead of creating new object for the same data 
a = 5
b = 5
print(a is b)
print(a is not b)
print(a == b)
print(id(a))
print(id(b))
```

1. Lists, dictionaries, and sets are mutable (their contents can be changed after creation). Python usually creates separate objects for separately created mutable objects, so is is generally False even if their values are the same.
2. Integers, floats, strings, tuples(only sometimes), booleans, and bytes are immutable (their contents cannot be changed after creation). Python may reuse (intern/cache) some immutable objects, so is may return True for objects with the same value, but this is an implementation optimization and should not be relied upon.

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   # True (same values)
print(a is b)   # False (different objects)
#"is" a points to the first list, while b points to the second list. They are two different objects stored in different places in memory.
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = a

print(a == b)   # True (same values)
print(a is b)   # true (same objects)
#"is" a points to the first list, while b points to the a or first list . so its same object and returns true 
```

```python
a = (1, 2, 3)
b = (1, 2, 3)

print(a == b)   # True
print(a is b)  # May be True or False
```

#### Membership operator 
- two types of membership operator 
1. in
2. not in
- membership operator is used to check the membership in a sequence (like string,list, tuples, sets , dict) , membership of characters , value or variable or number or substring.

```python
name ="Yeshwitha B M"
print("Y" in name )
print("y" in name ) # because in python almost everything is case sensitive 
print(" B M" in name )
print("y" not in name)
```
