### Variables

- Variables are used to store data in a computer's memory.
- When a variable is created, Python stores the value in memory.
- The variable name is used to access the stored value.
- Think of a variable as a label that points to a location in memory.

### Python Variable Name Rules

- A variable name must start with a letter or the underscore character
- A variable name cannot start with a number
- A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
- Variable names are case-sensitive (firstname, Firstname, FirstName and FIRSTNAME) are different variables.

#tip: A variable can have a short name (like x, y, z), but a more descriptive name (firstname, lastname, age, country) is highly recommended.


```python
first_name = "Yeshwitha" #variable to store first name 
age = "22" #variable to store age
country = "India" #variable to store country
print(first_name, age, country)  #print the data stored in the variables    
```

```python
#invalid variable name because it contains a hyphen and variable names cannot contain hyphens. 
# They can only contain letters, numbers, and underscores.
first-name="Yeshwitha" 
print(first-name)

#invalid variable examples : first-name, first@name, first$name, num-1, 1num
```

### naming conventions (case types) used in programming, especially in Python.

- student_name = "Alice"    # snake_case
- studentName = "Alice"     # camelCase
- StudentName = "Alice"     # PascalCase
- MAX_SIZE = 100            # UPPER_SNAKE_CASE
- student = "Alice"         # lowercase
- STUDENT = "Alice"         # UPPERCASE



```python
#Multiple variables can also be declared in one line:

first_name, age, country = "Yeshwitha", "22", "India"
print(first_name, age, country)  #print the data stored in the variables
```

### local variables in python 
Local variables are the variables that are created in a function or a class and exist only while these constructs are executing. These variables cannot be accessed outside the function/ class.

```python
def function_example():
    #local variable
    local_variable = "I am a local variable"
    print(local_variable)

function_example()
print(local_variable)  # This will raise an error because local_variable is out of function scope and cannot be accessed outside the function.
```

### Global variable 
These are the variables that are declared outside the functions/ classes. These will be available throughout the program.


```python
global_variable = "I am a global variable"  #global variable
def function_example():
    print(global_variable)  #global variable can be accessed inside the function
  
function_example()
print(global_variable)  #global variable can be accessed outside the function
```

```python
#del keyword is used to delete a variable in Python. Once a variable is deleted, it can no longer be accessed or used in the program.
name="Python"
print("Name=",name)
name = "PythonGeeks" # Reassigning the value
print("New Name=",name)
del name #Deleting the variable
print("Deleted",name)
```
