### adding elments to a list
- elements can be added to a list in different method 
#### append() method 
- a built-in method used to add an item at the end of a list 
- syntax is list.append(value) takes only 1 arguements 
#### insert() method 
- a built-in method used to add an item at a specific position.
- syntax is list.insert(position,value ) takes 2 arguements , position is index value to insert at specific position.
#### extend() method 
- a built-in method used to add all  items of one list to another list 
- syntax is list1.extend(list2)

````python
my_list.append(3) #takes only one arguement
my_list
my_list.insert(5,5)
my_list
````

````python
my_list=[1,2,3,4]
my_list3=[8,9,0]
my_list.extend(my_list3)
print(my_list)
print(my_list3)
````

````python
srt=[4,5,3,2,1]
srt1=[1.0,2.6,6,8.8,7]
srt2=["aii",'hello','what','really']
srt3=[True,False]
srt4=[1,'hi',45,True,5.0]
srt.sort()
srt1.sort()
srt2.sort()
srt3.sort()
print(srt)
print(srt1)
print(srt2)
print(srt3)
````

#### Reverse() function
- The reverse() method is used to reverse the order of elements in a list.
- It modifies the original list and returns None.
- syntax list.reverse()


````python
nums = [1, 2, 3, 4, 5]

nums.reverse()

print(nums)
````

````python
fruits = ["apple", "banana", "mango"]

fruits.reverse()

print(fruits)
````

````python
values = [True, False, True]

values.reverse()

print(values)
````

````python
nums = [1, 2, 3]

print(nums.reverse())
#reverse() changes the original list. It does not return a new list.
````

````python
nums = [4, 1, 5, 2]

nums.sort(reverse=True)

print(nums)
````

#### remove()
- The remove() method is used to remove the first occurrence of a specified value from a list.

- It modifies the original list and returns None.
- Syntax
list.remove(value)
- Parameter
value → The element you want to remove.


````python
nums = [10, 20, 30, 40]

nums.remove(30)

print(nums)
````

````python
fruits = ["apple", "banana", "mango"]

fruits.remove("banana")

print(fruits)
````

````python
#Example 3: Duplicate Elements remove() removes only the first occurrence.
nums = [1, 2, 3, 2, 4]

nums.remove(2)

print(nums)
````

````python
values = [True, False, True]

values.remove(True)

print(values)
````

````python
nums = [1, 2, 3]

nums.remove(5)
````

## `pop()`

Removes and **returns** an element from a list.

### Syntax

```python
list.pop(index)
```

- **index** *(optional)* → Position of the element to remove.
- If no index is provided, the **last element** is removed.

### Returns

- The removed element.

---

````python
nums = [10, 20, 30, 40]

x = nums.pop()

print(x)
print(nums)
````

````python
#pop by index
nums = [10, 20, 30, 40]

x = nums.pop(1)

print(x)
print(nums)
````

````python
#Pop the First Element
nums = [100, 200, 300]

x = nums.pop(0)

print(x)
print(nums)
````

````python
#Pop with Negative Index
nums = [10, 20, 30, 40]

print(nums.pop(-2))
print(nums)
````

````python
nums = []

nums.pop()
````

````python
nums = [1, 2, 3]

value = nums.pop()

print(value)
print(nums)
````
