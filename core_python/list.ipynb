### list 
- a collection of similar or different data types or items .
- a list is created by placing items inside [] separated by comas 
- an empty list can be created (num=[])  
- a list can have a duplicates 
- a list is a indexded (starts with 0 )
- items in a lists are mutable (changeable)
- example = [1,2,2,"d","hello",2.3]


```python
list=[1,2,3,4,4,"o",2.4555]
print(type(list))
print(list)
```

### accessing elements of a single dimensional list
- a single dimensional list is a list where elements are listed one after the other 
- each element is allotted a unique number called index .

```python
my_list=[1,2,3,4,"hello"]
my_list[3]
```

### negative indexing 
- negative indexing is allabout accessing elements from the last of the list.
- the positive indexes are from start like 0 1 2 3 ...,
- the last elements of any list has index starts from -1,-2,-3,-4., and so on .


```python
my_list[-2]
```

### Accessing elements of a multi-dimensional list
- multi-dimensional list is list containing another list.


```python
my_list1=[[1,2,3],"yeshu",[5,"hello",2.6]]
my_list1[1][2]
my_list1[-1][-2][-2]
```

```python
my_list3=[[1,2,3],[['a','b','c'],5,6]]
my_list3[1][0][1]
```

### adding elments to a list
- elements can be added to a list in different method 
#### append() method 
- a built-in method used to add an item at the end of a list 
- syntax is list.append(value) takes only 1 arguements 
#### insert() method 
- a built-in method used to add an item at a specific position.
- syntax is list.insert(position,value ) takes 2 arguements , position is index value to insert at specific position.
#### extend() method 
- a built-in method used to add all  items of one list to another list 
- syntax is list1.extend(list2)

```python
my_list.append(3) #takes only one arguement
my_list
my_list.insert(5,5)
my_list
```

```python
my_list=[1,2,3,4]
my_list3=[8,9,0]
my_list.extend(my_list3)
print(my_list)
print(my_list3)
```
