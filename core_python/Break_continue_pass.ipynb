# Loop Control Statements
- Loop control statements are used to alter the normal execution of a loop.
- these statements are used only inside loop body
- Python provides three loop control statements:
1. `break` – Terminates the loop immediately.
2. `continue` – Skips the current iteration and moves to the next.
3. `pass` – Does nothing; acts as a placeholder.

### Break
```text
            Entry Loop
                │
                ▼
       Test Loop Condition?
          ┌───────────────┐
     False│               │True
          ▼               ▼
    Exit From Loop    Is break?
                        │
                 ┌──────┴──────┐
              Yes│             │No
                 ▼             ▼
         Exit From Loop   Execute Remaining
                              Loop Body
                                  │
                                  ▼
                          Back to Condition
```

````python
count=1
while count<=10:
    print(count)
    count+=1
    if count==7:
        break
    print("Hi")
print("out from loop")
````

````python
list1=["hi","hello","welcome"]
list2=["krishna","ram","madhav"]
for item in list1:
    for name in list2:
        print(item,name)
        if item=="hello" and name=="ram":
            break # break is used within inner loop so out from inner loop 
    print("out from inner loop")
print("out from outer loop")
````

### Continue 
```text
            Entry Loop
                │
                ▼
       Test Loop Condition?
          ┌───────────────┐
     False│               │True
          ▼               ▼
    Exit From Loop   Is continue?
                        │
                 ┌──────┴──────┐
              Yes│             │No
                 ▼             |
      (skip next line )        ▼
      Go Back to Loop     Execute Remaining
          Condition           Loop Body
                 │                │
                 └──────┬─────────┘
                        ▼
               Back to Condition
```


````python
count=1
while count<=10:
    print(count)
    count+=1
    if count==7:
        print("skipped hi")
        continue
    print("Hi")
print("out from loop")
````

````python
list1=["hi","hello","welcome"]
list2=["krishna","ram","madhav"]
for item in list1:
    for name in list2:
        if item=="hello" and name=="ram":
            continue # continue is used within inner loop 
        print(item,name) 
    print("out from inner loop")
print("out from outer loop")
````

### Pass 
- pass is most commonly used as a placeholder for empty functions, classes, conditional blocks, exception handlers, or loops that you'll implement later.
- "I know something should go here, but I'm not writing it yet."
```text
            Entry Loop
                │
                ▼
       Test Loop Condition?
          ┌───────────────┐
     False│               │True
          ▼               ▼
    Exit From Loop     Is pass?
                        │
                 ┌──────┴──────┐
              Yes│             │No
                 ▼             ▼
        Do Nothing       Execute Code
                 │             │
                 └──────┬──────┘
                        ▼
               Remaining Loop Body
                        │
                        ▼
               Back to Condition
```


````python
for i in range(1,11):
    pass # prints nothing
````
