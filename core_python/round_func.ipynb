### round() Function
- The round() function is used to round a number to the nearest integer.
- The syntax is round(number, no of digits) it takes two parameters. by default if i don't give no of digits it will give to nearest of integer.


```python
print(round(7))
print(type(round(7)))
```

```python
print(round(7,2))
print(type(round(7,2)))
```

```python
print(round(7.63))
print(type(round(7.63)))
```

```python
print(round(2.34331,2))
print(type(round(2.34331,2)))
```

#### round(number, -n)

- When ndigits is negative, Python rounds the number to the left of the decimal point.

1. -1 → Nearest 10
2. -2 → Nearest 100
3. -3 → Nearest 1000
4. -4 → Nearest 10000

```python
print(round(674,-1))
print(round(674,-2))
print(round(674,-3))
print(round(674,-4))
print(round(674,-45))
```

```python
print(round(665,-1))
print(round(675,-1))
```

```python
print(round(676,-1))
```

```python
print(round(-8/3))
print(round(-1.5))
print(round(-8/3,2))
```
