#### Random Module (`random`)

The **`random`** module is a built-in Python module used to generate random numbers, randomly select elements, and shuffle sequences.

**Import**

```python
import random
```

---

#### `random.random()`

Returns a random floating-point number between **0.0** (inclusive) and **1.0** (exclusive).

**Syntax**

`random.random()`

---

#### `random.randint(start, end)`

Returns a random integer between the given range (**both start and end are included**).

**Syntax**

`random.randint(start, end)`

---

#### `random.randrange(start, stop, step)`

Returns a random integer from the specified range.

Works similarly to Python's `range()` function.

**Syntax**

`random.randrange(start, stop, step)`

---

#### `random.uniform(start, end)`

Returns a random floating-point number between the given range.

**Syntax**

`random.uniform(start, end)`

---

#### `random.choice(sequence)`

Returns **one random element** from a sequence such as a list, tuple, or string.

**Syntax**

`random.choice(sequence)`

---

#### `random.choices(sequence, k=n)`

Returns **multiple random elements** from a sequence.

Duplicates **are allowed**.

**Syntax**

`random.choices(sequence, k=n)`

---

#### `random.sample(sequence, k)`

Returns **multiple unique random elements** from a sequence.

Duplicates **are not allowed**.

**Syntax**

`random.sample(sequence, k)`

---

#### `random.shuffle(list)`

Randomly rearranges (shuffles) the elements of a list.

- Modifies the original list.
- Returns `None`.

**Syntax**

`random.shuffle(list)`

---

#### `random.seed(value)`

Initializes the random number generator.

Using the same seed value produces the same sequence of random numbers every time the program runs.

Useful for testing and reproducible results.

**Syntax**

`random.seed(value)`

---

#### Quick Summary

| Method | Description |
|---------|-------------|
| `random()` | Returns a random float between `0.0` and `1.0`. |
| `randint(a, b)` | Returns a random integer between `a` and `b` (inclusive). |
| `randrange()` | Returns a random integer from a specified range. |
| `uniform(a, b)` | Returns a random float between `a` and `b`. |
| `choice()` | Returns one random element from a sequence. |
| `choices()` | Returns multiple random elements (duplicates allowed). |
| `sample()` | Returns multiple unique random elements (no duplicates). |
| `shuffle()` | Randomly rearranges the elements of a list. |
| `seed()` | Produces the same random sequence for reproducible results. |
