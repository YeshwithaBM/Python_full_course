```python
print(len("string"))
```

```python
print(len(12345)) #gives an type error because len() function only takes "string" as an input.
```

```python
#str can concatinate only string not integers
length=len("yeshwitha B M")
print("My name has" + length +"characters")
```

#### Type Casting or Type conversion
- can convert type of data type .

```python
#now no error because we converted to string
length=len("yeshwitha B M")
print("My name has" + " " + str(length) + " "+"characters")
```

```python
#cannot convert the char as a int , but can convert number as string to int
print(int("45"))
print(int("hello"))
```

- we have : 
1. int() - to convert to int
2. float() - to convert to float 
3. str() - to convert to str 

```python
print(int(10.7888)) # convert float (10.78888) to  int (10) 
print(int("123"))   # convert string which has base 10 decimal (ex "123") and covert to int
print(float(10)) #convert int to float 
print(float("12")) # convert string to float
print(str(10)) #convert int to string
print(str(10.0)) #convert float to string
```

```python
print(type(int(10.7888))) # convert float (10.78888) to  int (10) 
print(type(int("123")))   # convert string which has base 10 decimal (ex "123") and covert to int
print(type(float(10))) #convert int to float 
print(type(float("12"))) # convert string to float
print(type(str(10))) #convert int to string
print(type(str(10.0))) #convert float to string
```
