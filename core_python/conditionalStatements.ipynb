#### Conditinal statements 
- these are decision making statements .
- because of this conditionals statements python is able to make a decision .

#### If statement (one way)
- if the condition is true , the line executes or if not it is skipped
- syntax : if condition: do this body  (usually in condition we use comparision operator)                       

#### if- else statement (two way)
- if the condition is true ,the line executes or if not it i skipped and moves to else part (usually else doesn't take any conditions defined, the other than if is executed )

#### if - elif statement (multi way)
- if the condition is true , the line executes or if not moves to elif condition and then if true executes or if not moves to else part.

##### Leap Year Rules

- A year is a leap year if:

1. It is divisible by 4.
2. But if it is divisible by 100, it is not a leap year.
3. However, if it is divisible by 400, it is a leap year.

```python
#leap year
year=int(input("Enter a number"))
if (year%4==0 and year%100!=4) or year%400==0 :
    print(f"{year} is a leap year")
else:
    print("Not a leap year")
```

```python
#Nested if and else 
year=int(input("Enter a number"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"{year} is a leap year")
        else:
            print("not a Leap year")
    else:
       print(f"{year} is a leap year")
else:
    print("Not a leap year")

```
