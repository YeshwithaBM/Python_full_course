### Functions 
- functions is a block of code where it does specific task when it is called .
- A **function** is a reusable block of code that performs a specific task.  
It is defined once and can be called multiple times, making programs cleaner and reducing code duplication.
##### Defining function
```python
def function_name(parameters):
    # Function body
    return statement
```
> **Note:** Parameters and the `return` statement are **optional**. A function can be created with or without them depending on the requirement.
##### Calling a Function
- A function is called by writing its **name followed by parentheses `()`**. When called, Python executes the code inside the function. If parameters are specified while defining function we need to pass parameters (we call arguements) while calling the function.

```python
function_name(arguements)
```
##### Types of Functions
1. Built-in Functions
Functions that are already provided by Python and can be used directly.
**Examples:**
```python
print("Hello")
len("Python")
input("Enter your name: ")
```
2. User-defined Functions
Functions created by the programmer using the `def` keyword to perform specific tasks.
**Example:**

```python
def greet():
    print("Hello!")

greet()
```

````python
def greet(): #defining the function
    print("hi")
    print("yeshwitha")

greet()#calling the function
greet()#calling the function
greet()#calling the function
````
