#### Tuples 
- A tuples is collection of data items stored in single variable (diff data types also can be  stored ).
- the syntax is 
``` tuples1=(1,2,3,4)``` enclosed with square brackets .
- can store diff data types or items ``` tuples2=(1,2,2.0,"hi",True)```
- tuples cannot be stored as single item without coma next to that for example {```tuples4=(10)``` is not a tuple }, {```tuple5=(10,)```} is valid.
- tuples cannot be modified so it is *Immutable*  
- can be acessed using index.
- slicing also possible in tuples 

````python
tuple1=(12,6,-3,"yeshu",True)
print(tuple1)
print(type(tuple1))
````

````python
#Indexing
print(tuple1[2]) # ususally for indexing we use [] to access the value
````

````python
tuple2=(10)
tuple3=(10,)
print(type(tuple2))
print(type(tuple3))
````

````python
#cannot modify ,its immutable ,so error is causing
tuple1[2]=9
print(tuple1)
````

````python
#duplicates are allowed 
tuple4=(1,2,2,2,3,4)
print(tuple4)
````

````python
#slicing 
print(tuple4[1:4])
print(tuple4[-1])
print(tuple4[1:6:2])
````

##### Difference Between List and Tuple

| Feature | List | Tuple |
|---------|------|-------|
| **Syntax** | `[]` | `()` |
| **Mutability** | Mutable (can be modified) | Immutable (cannot be modified) |
| **Performance** | Slightly slower | Faster than lists |
| **Memory Usage** | Uses more memory | Uses less memory |
| **Methods** | Many methods (`append()`, `extend()`, `remove()`, `pop()`, `sort()`, etc.) | Only 2 methods (`count()`, `index()`) |
| **Modification** | Elements can be added, removed, or updated | Elements cannot be added, removed, or updated after creation |
| **Hashable** | Not hashable | Hashable (if all elements are hashable), so it can be used as a dictionary key |
| **Use Case** | Data that changes frequently | Fixed data that should not change |
