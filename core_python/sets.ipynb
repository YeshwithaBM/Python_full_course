#### Sets in Python

A **set** is a built-in data type used to store **multiple unique elements**.

Sets are:

- Sets are enclosed with {} flower brackets 
- **Unordered** (elements have no fixed position or index).
- **Mutable** (elements can be added or removed).
- **Do not allow duplicate values.**
- Store only **immutable (hashable)** elements like integers, floats, strings, tuples, etc.


````python
set1={1,2,1,2,1,2,3,4,4,2} # does allow duplicates 
print(set1)
````

````python
set2={8,5,2,2,1,15,6} #unorder that means its doesnt return a value like in whch order i gave 
print(set2)
````

````python
set3={3,4,2,2,78665,0}
set3[4] #error because indexing is not posible
````

````python
set4={True,1,2,2,3,90,34,89,0,False,"yeshu"}
print(set4)
````

````python
## sets are immutable ,we cannot change 
## but complete set can be changes we have some methods 
set1={1,2,32,2,22}
set1.add(99)
print(set1)
print(len(set1))
````

````python
#remove an item
set1={1,2,32,2,22}
set1.remove(2)
print(set1)
print(len(set1))
````

````python
#remove an item
set1={1,2,32,2,22}
set1.remove(44) #raises an key error
print(set1)
print(len(set1))
````

````python
#discard an item ,it removes as well
set1={1,2,32,2,22}
set1.discard(2)
print(set1)
print(len(set1))
````

````python
#discard an item ,it removes as well
set1={1,2,32,2,22}
set1.discard(44) # it doesn't raises any error 
print(set1)
print(len(set1))
````

````python
#clear all the elements 

set1={1,2,32,2,22}
set1.clear()
print(set1)
print(len(set1))
````

````python
#pop will remove any element and returns as well

set1={1,2,32,2,22}
c=set1.pop() #takes no arguements
print(c)
print(set1)
print(len(set1))
````
