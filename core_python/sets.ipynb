#### Sets in Python

A **set** is a built-in data type used to store **multiple unique elements**.

Sets are:

- Sets are enclosed with {} flower brackets 
- **Unordered** (elements have no fixed position or index).
- **Mutable** (elements can be added or removed).
- **Do not allow duplicate values.**
- Store only **immutable (hashable)** elements like integers, floats, strings, tuples, etc.

#### Union Operation (`union()`)

The **Union** operation is used to **combine two or more sets into a single set**.
- All unique elements from each set are included.
- Duplicate elements are removed automatically.
##### Method
`set.union(*iterables)`
- **Parameters**:
- Accepts one or more **iterables**, such as:
  Sets,
  Lists
   ,Tuples
  ,Strings
- **Returns**: A new set containing all unique elements.
---
##### Union Operator (`|`)
The `|` operator is another way to perform the union operation.
- **Syntax** :
`set1 | set2`

- **Note**
1.  The `|` operator works **only with sets**.
2. It **does not** accept lists, tuples, strings, or other iterables.
3.  Using `|` with non-set objects raises a `TypeError`.

````python

set1={"yeshu","chai","ranju"}
set2={"suja","manju","yeshu"}
print(set1.union(set2)) # or we can pass set1 as arguements 
print(set2.union(set1))
print(set1 | set2 ) # with help of | operator we can do union
````

````python
lst=[1,5,2,2]
print(set1.union(lst)) # any iterables can pass as a arguements 
````

````python
print(set1 | lst) # but incase of | operator we cannot combine diff types like sets and list
````

````python
#Python's built-in set operations, the | operator is only meant to work with sets (or set-like types such as frozenset).
lst2=[3,4,5,6]
print(lst | lst2) 
````

### Update Operation (`update()`)

The **`update()`** method adds all elements from one or more **iterables** to an existing set.

- Modifies the original set.
- Duplicate values are ignored.
- Accepts sets, lists, tuples, strings, etc.
- Returns `None`.

**Syntax**

`set.update(*iterables)`

---

#### `union()` vs `update()`

| `union()` | `update()` |
|------------|------------|
| Returns a new set | Modifies the original set |
| Original set remains unchanged | Original set is updated |
| Returns a set | Returns `None` |
| Accepts any iterable | Accepts any iterable |

````python
set9={1,5,0}
set8={8,7}
print(set9.update(set8))#modifies the given set , here modified set 9
print(set9)
````

````python
set9={1,5,0}
set8={8,7}
print(set9.update(lst2))# takes anu iterable as a arguements 
print(set9)
````

### Intersection Operation (`intersection()`)

Returns a **new set containing only the common elements** from two or more iterables.

- Original set remains unchanged.
- Accepts any iterable (set, list, tuple, string, etc.).
- Returns a new set.

**Syntax:** `set.intersection(*iterables)`

#### `&` Operator

- Another way to perform intersection.
- Works **only with sets**.
- Returns a new set.

| `intersection()` | `&` |
|------------------|-----|
| Accepts any iterable | Accepts only sets |
| Returns a new set | Returns a new set |

````python
set10={9,8,6,0,8,6}
set6={6,8,3,2}
lst=[1,5,2,2]
print(set10.intersection(set6)) # cam pass any iterables as arguements
print(set10.intersection(lst)) # no common in both , (6 is duplicates not a common)
````

````python
print(set10 & set8) # operator to intersection of sets ,only takes sets .
print(set10 & set8 & set1) # there is only common in set10 and set8 so no common in set1 so returns empty set()
````

### Intersection Update (`intersection_update()`)

The **`intersection_update()`** method updates the original set by **keeping only the common elements** found in all iterables.

- Modifies the original set.
- Removes elements that are not common.
- Accepts any iterable (set, list, tuple, string, etc.).
- Returns `None`.

**Syntax:** `set.intersection_update(*iterables)`

---

#### `intersection()` vs `intersection_update()`

| `intersection()` | `intersection_update()` |
|------------------|-------------------------|
| Returns a new set | Modifies the original set |
| Original set remains unchanged | Original set is updated |
| Returns a set | Returns `None` |
| Accepts any iterable | Accepts any iterable |

````python
a={89,45,23,456}
b={75,45,26,23}
print(a.intersection_update(b)) # doesn't returns new set()
print(a) # modifies existing sets 
````

````python
a={789,123,456}
b={123}
print(a &= b) # no operators method is used for updation of intersection
print(a)
````

#### Difference Operation (`difference()`)

The **`difference()`** method returns a **new set containing elements that are present in the first set but not in the other iterable(s)**.

- Original set remains unchanged.
- Accepts any iterable.
- Returns a new set.

**Syntax:** `set.difference(*iterables)`

````python
set1={"yeshu","ranju","chai"}
set2={"manju","suja","yeshu"}
print(set1.difference(set2)) #it returns a new set contaning the elements present in first set1 but not other passed set2
````

#### Symmetric Difference (`symmetric_difference()`)

Returns a **new set containing elements that are present in either set, but not in both**. (set1 union of set2 - set1 intersection of set2 )

- Original set remains unchanged.
- Returns a new set.
- Accepts one iterable.

**Syntax:** `set.symmetric_difference(iterable)`

---

#### Symmetric Difference Update (`symmetric_difference_update()`)

Updates the original set by **keeping only the elements that are unique to each set**.

- Modifies the original set.
- Returns `None`.
- Accepts one iterable.

**Syntax:** `set.symmetric_difference_update(iterable)`

````python
set1={"yeshu","ranju","chai"}
set2={"manju","suja","yeshu"}
set3={"hello","ranju"}
# can pass multiple iterables as well 
print(set1.symmetric_difference(set2)) # common elements are not returned 
````

#### `isdisjoint()`
- Checks whether **two sets have no common elements**.
- Returns `True` if there are no common elements, otherwise `False`.
- **Operator:** None

---

#### `issubset()` (`<=`, `<`)
- Checks whether **all elements of the first set exist in the second set**.
- Returns `True` if the first set is a subset of the second.
- **Operators:** `<=` (subset), `<` (proper subset)

---

#### `issuperset()` (`>=`, `>`)
- Checks whether **the first set contains all elements of the second set**.
- Returns `True` if the first set is a superset of the second.
- **Operators:** `>=` (superset), `>` (proper superset)

````python
set1={"yeshu","ranju","chai"}
set2={"manju","suja","yeshu"}
set3={"hello","ranju"}
set4={"hello","what"}
print(set1.isdisjoint(set3))
print(set1.isdisjoint(set4))
````

````python
set1={"yeshu","ranju"}
set2={"manju","suja","yeshu"}
set3={"hello","ranju"}
set4={"ranju","yeshu","chai"}
print(set1.issubset(set2))
print(set1.issubset(set4))
````

#### Frozen Set (`frozenset`)

A **frozenset** is an **immutable version of a set**.

- Stores unique, unordered elements.
- Cannot be modified after creation.
- Supports only non-modifying methods.

**Syntax:** `frozenset(iterable)`

**Methods:** `copy()`, `union()`, `intersection()`, `difference()`, `symmetric_difference()`, `isdisjoint()`, `issubset()`, `issuperset()`

````python
set1={"yeshu","ranju"}
set2={"manju","suja","yeshu"}
set3=frozenset(set2)
print(set2)
set3.add(99)



````
