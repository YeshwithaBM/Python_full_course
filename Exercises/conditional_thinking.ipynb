- check if a given year is a leap year 

```python
year=int(input())
if (year%4==0 and year%100!=0) or year%400==0:
    print("leap year")
else:
    print("its not a leap year")
```

```python
lst=[[1,2,3],
     [1,2,3],
     [1,2,3]]
user=int(input("enter position "))
if user==11:
    lst[0][0]="x"
    print(lst)
elif user==12:
    lst[0][1]="x"
    print(lst)
elif user==13:
    lst[0][2]="x"
    print(lst)
elif user==21:
    lst[1][0]="x"
    print(lst)
elif user==22:
    lst[1][1]="x"
    print(lst)
elif user==23:
    lst[1][2]="x"
    print(lst)
elif user==31:
    lst[2][0]="x"
    print(lst)
elif user==32:
    lst[2][1]="x"
    print(lst)
else:
    lst[2][2]="x"
    print(lst)
```
