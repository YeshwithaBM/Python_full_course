- Take a number and print whether its positive ,negative or zero


```python
number=int(input())
if number>0:
    print("Positive number")
elif number<0:
    print("Negative number")
else:
    print("Zero")
```

- check if a number is even or odd


```python
number=int(input())
if number==0:
    print("zero")
elif number%2==0:
    print("even number")
else:
    print("odd")
```

- check if a number is divisible by 5


```python
number=int(input())
if number%5==0:
    print(True)
else: 
    print(False)
```

- check if a number is divisible by both 3 and 5


```python
num=int(input())
if num%3==0 and num%5==0:
    print(True)
else:
    print(False)
```

- check if a given year is a leap year 

```python
year=int(input())
if (year%4==0 and year%100!=0) or year%400==0:
    print("leap year")
else:
    print("its not a leap year")
```

```python
char=input()
vowels=['a','e','i','o','u']
if char in vowels:
    print("vowels")
else:
    print("consonant")
```

```python
char=input()
if char.isupper():
    print("upper case")
elif char.islower():
    print("lower case")
elif char.isdigit():
    print("digit")
else:
    print("special characters")
```
