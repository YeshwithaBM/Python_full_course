#### Linked List

##### Definition
A Linked List is a linear data structure made of **nodes**, where each node stores data and a reference to the next node.

##### Why Linked List?
- Arrays require shifting elements during insertion and deletion.
- Linked Lists avoid shifting by changing the links between nodes.
- Dynamic in size and does not require contiguous memory.

##### Node
A node is the basic building block of a Linked List.

Each node has:
- `data` → Stores the value.
- `next` → Stores the reference to the next node.

##### Head
- `head` is a reference to the first node of the Linked List.
- If `head` is lost, the entire Linked List becomes inaccessible.

##### Characteristics
- Nodes are connected using references (`next`).
- Elements are not stored in contiguous memory.
- Grows dynamically by creating new nodes.

##### Advantages
- Efficient insertion and deletion (no shifting).
- Dynamic size.
- Memory allocated as needed.

##### Disadvantages
- No direct indexing (`list[i]` is not possible).
- Sequential traversal is required.
- Extra memory is needed to store references.

##### Time Complexity

| Operation | Complexity |
|-----------|------------|
| Access by Index | O(n) |
| Search | O(n) |
| Insert at Beginning | O(1) |
| Insert at End* | O(n) |
| Delete at Beginning | O(1) |
| Delete at End | O(n) |
| Traversal | O(n) |

> *Insert at End becomes **O(1)** if a tail pointer is maintained.

### Traversal

Traversal is the process of visiting each node in a Linked List exactly once, starting from the head and following the `next` reference until `None` is reached.

```python
temp = head

while temp is not None:
    print(temp.data)
    temp = temp.next
```

### Search in Linked list

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
    def search(head,target):
        temp=head
        while temp is not None:
            if temp.data==target:
                return True
            temp=temp.next
        return False
first=Node(10)
second=Node(20)
third=Node(30)
first.next=second
second.next=third
finding=Node.search(first,30)
print(finding)
```

### Insert at beginning

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
first=Node(10)
second=Node(20)
third=Node(30)
first.next=second
second.next=third
head=first
def insert_beginning(head,data):
    new_node=Node(data)
    new_node.next=head
    head=new_node
    return head.data
insert_beginning(first,5)
```

### Insert at end 

```python
def insert_end(head,data):
    new_node=Node(data)
    if head is None:
        return new_node
    temp=head
    while temp.next is not None:
        temp=temp.next
    temp.next=new_node
    return head
print(insert_end(first,40))


```

### Insert at a Specific Position

```python
def insert_position(head,data,position):
    new_node=Node(data)
    if position == 1:
        new_node.next = head
        head = new_node
        return head
    temp=head
    for i in range(position-2):
        temp=temp.next
    new_node.next = temp.next
    temp.next = new_node
    return head
insert_position(first,15,2)
    
```
