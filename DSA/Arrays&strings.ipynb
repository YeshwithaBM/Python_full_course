### Arrays 


```python
#array (dynamic array in python is basically a List )
A =[1,2,3,4]
print(A)
```

```python
#append - Insert element at the end of array - Time complexity is On average : O(1)*
A.append(5)
print(A)
```

```python
#Pop - Deleting element at the end of the array - O(1)
A.pop()
print(A)
```

```python
# Insert (not at the end of the array) - O(n) because we might has to shift many other elements.
A.insert(2,0)
print(A)
```

```python
#Modify an element - O(1)
A[2]=8
print(A)
```

```python
#Accessing an element given index - O(1)
print(A[3])
```

```python
#search through array or check if an element present so its is O(n)- because it goes on checking all or n elements .
if 5 in A:
    print(True)
```

```python
#to check the length of an array - O(1)
print(len(A))
```

```python
class Solution:
    def twoSum(self, nums, target):

        seen = {}

        for i, num in enumerate(nums):
            complement = target - num

            if complement in seen:
                return [seen[complement], i]

            seen[num] = i
s=Solution()
s.twoSum([11,2,15,7,8,1],9)
```

```python
class Solution:
    def removeDuplicates(self, nums):

        if not nums:
            return 0

        k = 1

        for i in range(1, len(nums)):
            if nums[i] != nums[k - 1]:
                nums[k] = nums[i]
                k += 1

        return k
s=Solution()
s.removeDuplicates([1,1,2])
```
