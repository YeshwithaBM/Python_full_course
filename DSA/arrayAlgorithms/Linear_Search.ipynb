### Linear search


```python
#Return the index of the target element.
def search(lst,target):
    for i in range(len(lst)):
        if lst[i]==target:
           return i
    else:
        return False
s=search([10,23,34,23,656,7,35],23)
s

       
```

```python
#Return True or False if the target exists.
def search(lst,target):
    for i in range(len(lst)):
        if lst[i]==target:
            return True
    else:
        return False
s=search([10,23,34,656,7,35],13)
s
```

```python
#Count how many times a target appears in an array.
def search(lst,target):
    count=0
    for i in range(len(lst)):
        if lst[i]==target:
            count+=1
    return count      
s=search([10,23,34,656,7,35,23],23)
print(s)

```

```python
#Find the first occurrence of a target.
def search(lst,target):

    for i in range(len(lst)-1,-1,-1):

        if lst[i]==target:
            return i

    return -1   
s=search([10,23,34,656,7,35,23],23)
print(s)

```

```python
#Find the last occurrence of a target.
def search(lst,target):

    for i in range(len(lst)):

        if lst[i]==target:
            return i

    return -1 
s=search([10,23,34,656,7,35,23],23)
print(s)
```
