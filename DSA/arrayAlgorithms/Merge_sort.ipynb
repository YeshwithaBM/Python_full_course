```python
#4. Index-Based Merge Sort (Interview Version)
def merge(nums, low, mid, high):

    temp = []

    i = low
    j = mid + 1

    while i <= mid and j <= high:

        if nums[i] <= nums[j]:
            temp.append(nums[i])
            i += 1
        else:
            temp.append(nums[j])
            j += 1

    while i <= mid:
        temp.append(nums[i])
        i += 1

    while j <= high:
        temp.append(nums[j])
        j += 1

    for k in range(low, high + 1):
        nums[k] = temp[k - low]


def merge_sort(nums, low, high):

    if low >= high:
        return

    mid = (low + high) // 2

    merge_sort(nums, low, mid)

    merge_sort(nums, mid + 1, high)

    merge(nums, low, mid, high)


nums = [8, 3, 6, 2]

merge_sort(nums, 0, len(nums) - 1)

print(nums)
```

```python
#3. Compact Recursive Merge Sort
def merge_sort(nums):

    if len(nums) <= 1:
        return nums

    mid = len(nums) // 2

    return merge(
        merge_sort(nums[:mid]),
        merge_sort(nums[mid:])
    )


def merge(left, right):

    result = []

    i = j = 0

    while i < len(left) and j < len(right):

        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


nums = [8,3,6,2]

print(merge_sort(nums))
```

```python
#1. Recursive Merge Sort (Using Slicing)
def merge_sort(nums):

    if len(nums) <= 1:
        return nums

    mid = len(nums) // 2

    left = merge_sort(nums[:mid])
    right = merge_sort(nums[mid:])

    return merge(left, right)


def merge(left, right):

    result = []

    i = j = 0

    while i < len(left) and j < len(right):

        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


nums = [8, 3, 6, 2]

print(merge_sort(nums))
```

```python
#5. Bottom-Up (Iterative) Merge Sort
def merge(nums, left, mid, right):

    temp = []

    i = left
    j = mid

    while i < mid and j < right:

        if nums[i] <= nums[j]:
            temp.append(nums[i])
            i += 1
        else:
            temp.append(nums[j])
            j += 1

    while i < mid:
        temp.append(nums[i])
        i += 1

    while j < right:
        temp.append(nums[j])
        j += 1

    nums[left:right] = temp


def merge_sort(nums):

    n = len(nums)

    size = 1

    while size < n:

        left = 0

        while left < n:

            mid = min(left + size, n)
            right = min(left + 2 * size, n)

            merge(nums, left, mid, right)

            left += 2 * size

        size *= 2


nums = [8,3,6,2]

merge_sort(nums)

print(nums)
```
