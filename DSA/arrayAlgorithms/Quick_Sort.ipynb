#### Quick Sort
##### Lomuto Partition Algorithm

1. Choose the last element as the pivot.

2. Initialize:

```text
i = low - 1
```

3. Traverse the array using `j` from `low` to `high - 1`.

4. If `arr[j] < pivot`:
   - Increment `i`.
   - Swap `arr[i]` and `arr[j]`.

5. After the loop ends:
   - Swap `arr[i + 1]` with the pivot.

6. Return `i + 1` (the pivot's final index).

````python
def partition(nums,low,high):
    pivot=nums[high]
    i=low-1
    for j in range(low,high):
        if nums[j]<pivot:
            i+=1
            nums[i],nums[j]=nums[j],nums[i]
    nums[i+1],nums[high]=nums[high],nums[i+1]
    return i+1

def quick_sort(nums,low,high):
    if low<high:
        pivot_index=partition(nums,low,high)
        quick_sort(nums,low,pivot_index-1)
        quick_sort(nums,pivot_index+1,high)
    return nums

nums=[8,3,6,2]
print(quick_sort(nums,0,len(nums)-1))
````

##### Median of Three Quick Sort

````python
def median_of_three(nums, low, high):

    mid = (low + high) // 2

    if nums[low] > nums[mid]:
        nums[low], nums[mid] = nums[mid], nums[low]

    if nums[low] > nums[high]:
        nums[low], nums[high] = nums[high], nums[low]

    if nums[mid] > nums[high]:
        nums[mid], nums[high] = nums[high], nums[mid]

    nums[mid], nums[high] = nums[high], nums[mid]


def partition(nums, low, high):

    median_of_three(nums, low, high)

    pivot = nums[high]

    i = low - 1

    for j in range(low, high):

        if nums[j] < pivot:

            i += 1

            nums[i], nums[j] = nums[j], nums[i]

    nums[i + 1], nums[high] = nums[high], nums[i + 1]

    return i + 1


def quick_sort(nums, low, high):

    if low < high:

        pivot_index = partition(nums, low, high)

        quick_sort(nums, low, pivot_index - 1)

        quick_sort(nums, pivot_index + 1, high)

    return nums


nums = [8,3,6,2,9,1,5]

print(quick_sort(nums,0,len(nums)-1))
````

### Tail Recursion Optimized Quick Sort


````python
def partition(nums, low, high):

    pivot = nums[high]

    i = low - 1

    for j in range(low, high):

        if nums[j] < pivot:

            i += 1

            nums[i], nums[j] = nums[j], nums[i]

    nums[i + 1], nums[high] = nums[high], nums[i + 1]

    return i + 1


def quick_sort(nums, low, high):

    while low < high:

        pivot_index = partition(nums, low, high)

        if pivot_index - low < high - pivot_index:

            quick_sort(nums, low, pivot_index - 1)

            low = pivot_index + 1

        else:

            quick_sort(nums, pivot_index + 1, high)

            high = pivot_index - 1

    return nums


nums = [111,3,6,2,3,1,5]

print(quick_sort(nums,0,len(nums)-1))
````

````python
#K-th Smallest Element (Quick Select)
#Quick Select is based on Quick Sort but recursively explores only one side of the partition.
def partition(nums, low, high):

    pivot = nums[high]

    i = low - 1

    for j in range(low, high):

        if nums[j] < pivot:

            i += 1

            nums[i], nums[j] = nums[j], nums[i]

    nums[i + 1], nums[high] = nums[high], nums[i + 1]

    return i + 1


def quick_select(nums, low, high, k):

    if low <= high:

        pivot_index = partition(nums, low, high)

        if pivot_index == k:

            return nums[pivot_index]

        elif pivot_index > k:

            return quick_select(nums, low, pivot_index - 1, k)

        else:

            return quick_select(nums, pivot_index + 1, high, k)


nums = [8,3,6,2,5]

k = 0          # 3rd smallest (0-based index)

print(quick_select(nums,0,len(nums)-1,k))
````
