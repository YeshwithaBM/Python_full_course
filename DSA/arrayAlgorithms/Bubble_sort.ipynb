### Sorting Algorithms
#### Bubble sort

```python
def bubble_sort(nums):
    n=len(nums)
    for i in range(n-1):
        for j in range(n-1-i):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
    return nums
nums=[5,2,4,1]
print(bubble_sort(nums))
```

```python
#Modify Bubble Sort to stop early if the array is already sorted (this is the optimized version)
def bubble_sort(nums):
    n=len(nums)
    for i in range(n-1):
        swapped = False
        for j in range(n-1-i):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                swapped = True
        if not swapped:
            break
    return nums
nums=[1]
print(bubble_sort(nums))
```

```python
#Sort in descending order.
def bubble_sort(nums):
    n=len(nums)
    for i in range(n-1):
        swapped = False
        for j in range(n-1-i):
            if nums[j]<nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                swapped = True
        if not swapped:
            break
    return nums
nums=[5,8,1,3,4]
print(bubble_sort(nums))
```

```python
#Count the number of swaps performed.
def bubble_sort(nums):
    n=len(nums)
    count=0
    for i in range(n-1):
        swapped = False
        for j in range(n-1-i):
            if nums[j]>nums[j+1]:
                nums[j],nums[j+1]=nums[j+1],nums[j]
                swapped = True
                count+=1
        if not swapped:
            break
    return count
nums=[5,4,3,2,1]
print(bubble_sort(nums))
```
