### Remove Element

```python

def removeElement(nums,val) :
        if not nums:
            return 0
        left=0
        for right in range(len(nums)):
            if nums[right]!=val:
                nums[left]=nums[right]
                left+=1
        return left
nums=[3,2,6,7,8,6,9,6,5,4,0,2,2,3]
k=removeElement(nums,3)
print(k)
print(nums[:k])
        
```
