### Remove Duplicates 


```python
def removeDuplicates(nums):
        if not nums:
            return 0
        left=0
        for right in range(1,len(nums)):
            if nums[left]!=nums[right]:
                left+=1
                nums[left]=nums[right]
        return left+1
nums = [1, 1, 2]
k = removeDuplicates(nums)
print(nums[:k])
```

```python
def removeElement( nums,val):
        if not nums:
            return 0
        left=0
        for right in range(0,len(nums)-1):
            if right!=val:
                nums[left]=nums[right]
                left+=1
        return left+1
nums=[3,2,2,3]
print(removeElement(nums,3))
```
