### Move zeros

```python
def moveZeros(nums) :
        if not nums:
            return 0
        left=0
        for right in range(len(nums)):
            if nums[right]!=0:
                nums[left]=nums[right]
                left+=1
        while left<len(nums):
             nums[left]=0
             left+=1
nums=[0,10,2,3,12,0]
print(moveZeros(nums))
print(nums)
```
