### Two Sum II

```python
# normal implementation to return values/ target where we found by adding
def two_sum(nums, target):
    left = 0
    right = len(nums) - 1

    while left < right:
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            return [nums[left], nums[right]]

        if current_sum > target:
            right -= 1
        else:
            left += 1       
    
nums=[1,3,4,6,8,11]
target=10
print(two_sum(nums,target))
```

```python
# if we have to return indices 
def two_sum(nums, target):
    left = 0
    right = len(nums) - 1

    while left < right:
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            return [left, right]

        if current_sum > target:
            right -= 1
        else:
            left += 1       
    
nums=[1,3,4,6,8,11]
target=10
print(two_sum(nums,target))
```

```python
## if we have to return indices 1-based (not considering from 0 but from 1 )
def two_sum(nums, target):
    left = 0
    right = len(nums) - 1

    while left < right:
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            return [left+1, right+1]

        if current_sum > target:
            right -= 1
        else:
            left += 1       
    
nums=[1,3,4,6,8,11]
target=10
print(two_sum(nums,target))
```
