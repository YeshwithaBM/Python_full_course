### Palindrome

```python
#abcdcba , racecar, abc
# I did 
def palindrome(string):
    left=0
    right = len(string) - 1
    while left < right:
        if string[left]==string[right]:
            left+=1
            right-=1
        else:
            return "Not a Palindrome"

    return "Its a palindrome"
string="aba"
print(palindrome(string))
```

```python
#better because stops at first iteration if elements doesn't match
def palindrome(string):

    left = 0
    right = len(string)-1

    while left < right:

        if string[left] != string[right]:
            return "Not a palindrome"

        left += 1
        right -= 1

    return "Its a Palindrome"
string="aba"
print(palindrome(string))
```

```python
#Ignore spaces ("nurses run" → True).

def palindrome(string):

    left = 0
    right = len(string) - 1

    while left < right:

        while left < right and string[left] == " ":
            left += 1

        while left < right and string[right] == " ":
            right -= 1

        if string[left] != string[right]:
            return "Not a Palindrome"

        left += 1
        right -= 1

    return "Its a palindrome"


string = "nurses run"

print(palindrome(string))
```

```python
#Ignore spaces and special characters ("A man, a plan, a canal: Panama" → True).
#it is not efficient , because of extra space used by "cleaned" space is o(n) 
import re
def palindrome(string):
    cleaned= re.sub(r'[^a-zA-Z0-9]', '', string)
    left=0
    right = len(cleaned) - 1
    while left < right:
        if cleaned[left]==cleaned[right]:
            left+=1
            right-=1
        else:
            return "Not a Palindrome"

    return "Its a palindrome"
string="A man, a plan, a canal: Panama"
print(palindrome(string.lower()))
```

```python
#Optimized of if special char and spaces present
#o(1) of space . time is as same as others i.e o(n)
def palindrome(string):

    left = 0
    right = len(string) - 1

    while left < right:

        while left < right and not string[left].isalnum():
            left += 1

        while left < right and not string[right].isalnum():
            right -= 1

        if string[left].lower() != string[right].lower():
            return False

        left += 1
        right -= 1

    return True


string = "A man, a plan, a canal: Panama"

print(palindrome(string))
```
