### Sort Array By Parity

```python
def sortByParity(nums) :
        if not nums:
            return 0
        left=0
        for right in range(len(nums)):
            if nums[right]%2==0:
                nums[left],nums[right]=nums[right],nums[left]
                left+=1
        return nums
nums=[3,1,2,5,0,74,9,3,5,4,8,12]
print(sortByParity(nums))

```
