### Problem Statement

- Reverse the given string in-place using the Two Pointer technique.

```python
def reverse_string(string):
    left=0
    right = len(string) - 1
    while left < right:
         string[left], string[right] = string[right], string[left]
         left+=1
         right-=1
    return string
string=["y","e","s","h"]
print(reverse_string(string))
```
