### Number of Sub-arrays of size K and average greater than or equal to Threshold 

#### Brute-Force 

```python
def numOfSubarrays(nums, k, threshold):

    count = 0
    left = 0
    right = k - 1

    while right < len(nums):
        window_sum = 0
        for i in range(left,right+1):
            window_sum+=nums[i]
        avg=window_sum/k
        if avg>=threshold:
            count+=1
        left += 1
        right += 1

    return count
nums = [2,2,2,2,5,5,5,8]
print(numOfSubarrays(nums,3,4))
```

### Optimized

```python
def numOfSubarraysThreshold(nums,k,threshold):
    window_sum=sum(nums[:k])
    count=0
    if window_sum>=threshold*k:
        count+=1
    for right in range(k,len(nums)):
        window_sum=window_sum-nums[right-k]+nums[right]
        if window_sum>=threshold*k:
            count+=1
    return count
nums=[2,2,2,2,5,5,5,8]
print(numOfSubarraysThreshold(nums,3,4))

```
