### Maximum Number of Vowels in a Substring of Given Length

### brute force

```python
def SubString(nums,k):
    max_count=0
    vowels="aeiouAEIOU"  
    for i in range(len(nums)-k+1):
        count=0
        right=i+k-1
        for j in range(i,right+1):
            if nums[j] in vowels:
                count+=1
        max_count=max(max_count,count)
    return max_count
nums="abciiidef"
print(SubString(nums,3))
    
```

### Optimized

```python
def maxVowels(s, k):
    vowels = "aeiouAEIOU"

    # Count vowels in first window
    vowel_count = 0
    for i in range(k):
        if s[i] in vowels:
            vowel_count += 1

    max_count = vowel_count

    # Slide the window
    for right in range(k, len(s)):

        # Remove left character
        if s[right - k] in vowels:
            vowel_count -= 1

        # Add right character
        if s[right] in vowels:
            vowel_count += 1

        max_count = max(max_count, vowel_count)

    return max_count


s = "abciiidef"
k = 3

print(maxVowels(s, k))
```
