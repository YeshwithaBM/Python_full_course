#### Maximum Average Subarray

##### Brute force approach

```python
def findMaxAverage(nums, k):

    max_average = float('-inf')

    # Loop through every possible window
    for start in range(len(nums) - k + 1):

        window_sum = 0

        # Calculate the sum of the current window
        for i in range(start, start + k):
            window_sum += nums[i]

        average = window_sum / k

        if average > max_average:
            max_average = average

    return max_average


nums = [1, 12, -5, -6, 50, 3]
k = 4

print(findMaxAverage(nums, k))
```

```python
def MaximumAvg(nums, k):
    max_avg = float("-inf")

    left = 0
    right = k - 1

    # Calculate first window sum
    window_sum = 0
    for i in range(k):
        window_sum += nums[i]

    while right < len(nums):

        avg = window_sum / k

        if avg > max_avg:
            max_avg = avg

        # No more windows left
        if right == len(nums) - 1:
            break

        # Slide the window
        window_sum = window_sum - nums[left] + nums[right + 1]

        left += 1
        right += 1

    return max_avg


nums = [10,20,30,40,50,60]
print(MaximumAvg(nums,3))
```

```python
def MaximumAvg(nums, k):
    # Sum of the first window
    window_sum = sum(nums[:k])

    max_avg = window_sum / k

    # right is the index of the new element entering the window
    for right in range(k, len(nums)):

        # Remove left element, add new right element
        window_sum = window_sum - nums[right - k] + nums[right]

        avg = window_sum / k

        if avg > max_avg:
            max_avg = avg

    return max_avg


nums = [10, 20, 30, 40, 50, 60]
print(MaximumAvg(nums, 3))
```
